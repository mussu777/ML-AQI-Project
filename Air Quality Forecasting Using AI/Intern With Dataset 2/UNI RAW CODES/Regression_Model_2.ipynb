```python
import pandas as pd
from darts import TimeSeries
```

```python
df = pd.read_csv("MY1_2023.csv")

df
```

```python
df1 = df
df1['Datetime'] = pd.to_datetime(df['Datetime'])

df1 = df.set_index('Datetime')
#df1
df2=df1.resample('1H').mean()
df2.reset_index(inplace = True)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.interpolate(method='bfill', inplace=True)
df2
```

```python
print(df2.isnull().sum())
```

```python
df = df2

target1 = 'field1'
test_l = 168
train_l = len(df) - test_l
date = 'Datetime'
# Create a TimeSeries, specifying the time and value columns
series1 = TimeSeries.from_dataframe(df, date, target1)

# Set aside the last 24 hours as a validation series
train1, val1 = series1[:train_l], series1[train_l:]

target2 = 'field2'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series2 = TimeSeries.from_dataframe(df, date, target2)

# Set aside the last 24 hours as a validation series
train2, val2 = series2[:train_l], series2[train_l:]

target3 = 'field3'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series3 = TimeSeries.from_dataframe(df, date, target3)

# Set aside the last 24 hours as a validation series
train3, val3 = series3[:train_l], series3[train_l:]

target4 = 'field4'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series4 = TimeSeries.from_dataframe(df, date, target4)

# Set aside the last 24 hours as a validation series
train4, val4 = series4[:train_l], series4[train_l:]

target7 = 'field7'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series7 = TimeSeries.from_dataframe(df, date, target7)

# Set aside the last 24 hours as a validation series
train7, val7 = series7[:train_l], series7[train_l:]

target8 = 'field8'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series8 = TimeSeries.from_dataframe(df, date, target8)

# Set aside the last 24 hours as a validation series
train8, val8 = series8[:train_l], series8[train_l:]
```

# Regression Model

```python
from darts.metrics import rmse, mae, mape
from darts.models import RegressionModel

# Train the regression model for target1
regression_model1 = RegressionModel(lags=1)
regression_model1.fit(train1)

# Test the regression model for target1
regression_predictions1 = regression_model1.predict(len(val1))

# Predict using the regression model for target1
test_predictions1 = regression_model1.predict(len(val1) + test_l)

# Calculate RMSE and MAE for target1
rmse1 = rmse(regression_predictions1, val1)
mae1 = mae(regression_predictions1, val1)
mape1 = mape(regression_predictions1, val1)

# Train the regression model for target2
regression_model2 = RegressionModel(lags=1)
regression_model2.fit(train2)

# Test the regression model for target2
regression_predictions2 = regression_model2.predict(len(val2))

# Predict using the regression model for target2
test_predictions2 = regression_model2.predict(len(val2) + test_l)

# Calculate RMSE and MAE for target2
rmse2 = rmse(regression_predictions2, val2)
mae2 = mae(regression_predictions2, val2)
mape2 = mape(regression_predictions2, val2)

# Train the regression model for target3
regression_model3 = RegressionModel(lags=1)
regression_model3.fit(train3)

# Test the regression model for target3
regression_predictions3 = regression_model3.predict(len(val3))

# Predict using the regression model for target3
test_predictions3 = regression_model3.predict(len(val3) + test_l)

# Calculate RMSE and MAE for target3
rmse3 = rmse(regression_predictions3, val3)
mae3 = mae(regression_predictions3, val3)
mape3 = mape(regression_predictions3, val3)

# Train the regression model for target4
regression_model4 = RegressionModel(lags=1)
regression_model4.fit(train4)

# Test the regression model for target4
regression_predictions4 = regression_model4.predict(len(val4))

# Predict using the regression model for target4
test_predictions4 = regression_model4.predict(len(val4) + test_l)

# Calculate RMSE and MAE for target4
rmse4 = rmse(regression_predictions4, val4)
mae4 = mae(regression_predictions4, val4)
mape4 = mape(regression_predictions4, val4)

# Train the regression model for target7
regression_model7 = RegressionModel(lags=1)
regression_model7.fit(train7)

# Test the regression model for target7
regression_predictions7 = regression_model7.predict(len(val7))

# Predict using the regression model for target7
test_predictions7 = regression_model7.predict(len(val7) + test_l)

# Calculate RMSE and MAE for target7
rmse7 = rmse(regression_predictions7, val7)
mae7 = mae(regression_predictions7, val7)
mape7 = mape(regression_predictions7, val7)

# Train the regression model for target8
regression_model8 = RegressionModel(lags=1)
regression_model8.fit(train8)

# Test the regression model for target8
regression_predictions8 = regression_model8.predict(len(val8))

# Predict using the regression model for target8
test_predictions8 = regression_model8.predict(len(val8) + test_l)

# Calculate RMSE and MAE for target8
rmse8 = rmse(regression_predictions8, val8)
mae8 = mae(regression_predictions8, val8)
mape8 = mape(regression_predictions8, val8)

# Create a DataFrame to store the results
results = pd.DataFrame({
    'Target': ['target1', 'target2', 'target3', 'target4', 'target7', 'target8'],
    'RMSE': [rmse1, rmse2, rmse3, rmse4, rmse7, rmse8],
    'MAE': [mae1, mae2, mae3, mae4, mae7, mae8],
    'MAPE': [mape1, mape2, mape3, mape4, mape7, mape8]
})

# Save the results to a CSV file
results.to_csv('RegressionModel_2.csv', index=False)
```
