```python
import pandas as pd
from darts import TimeSeries
```

```python
df = pd.read_csv("MY1_2023.csv")

df
```

```python
df1 = df
df1['Datetime'] = pd.to_datetime(df['Datetime'])

df1 = df.set_index('Datetime')
#df1
df2=df1.resample('1H').mean()
df2.reset_index(inplace = True)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.interpolate(method='ffill', inplace=True)
df2
```

```python
print(df2.isnull().sum())
```

```python
df = df2

target1 = 'field1'
test_l = 168
train_l = len(df) - test_l
date = 'Datetime'
# Create a TimeSeries, specifying the time and value columns
series1 = TimeSeries.from_dataframe(df, date, target1)

# Set aside the last 24 hours as a validation series
train1, val1 = series1[:train_l], series1[train_l:]

target2 = 'field2'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series2 = TimeSeries.from_dataframe(df, date, target2)

# Set aside the last 24 hours as a validation series
train2, val2 = series2[:train_l], series2[train_l:]

target3 = 'field3'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series3 = TimeSeries.from_dataframe(df, date, target3)

# Set aside the last 24 hours as a validation series
train3, val3 = series3[:train_l], series3[train_l:]

target4 = 'field4'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series4 = TimeSeries.from_dataframe(df, date, target4)

# Set aside the last 24 hours as a validation series
train4, val4 = series4[:train_l], series4[train_l:]

target7 = 'field7'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series7 = TimeSeries.from_dataframe(df, date, target7)

# Set aside the last 24 hours as a validation series
train7, val7 = series7[:train_l], series7[train_l:]

target8 = 'field8'
test_l = 168
train_l = len(df) - test_l

# Create a TimeSeries, specifying the time and value columns
series8 = TimeSeries.from_dataframe(df, date, target8)

# Set aside the last 24 hours as a validation series
train8, val8 = series8[:train_l], series8[train_l:]
```

# NLinear Model

```python
# importing model
from darts.models import NLinearModel

# creating a model object
modelnl1 = NLinearModel(input_chunk_length=24 , output_chunk_length=12, shared_weights=False, const_init=True, normalize=False, use_static_covariates=True)
# fitting the model
modelnl1.fit(train1, verbose = True)

# imports
from darts.metrics import mape
prednl1 = modelnl1.predict(n = len(val1), series = train1)
print("Mape = {:.2f}%".format(mape(train1 , prednl1)))

# creating a model object
modelnl2 = NLinearModel(input_chunk_length=24 , output_chunk_length=12, shared_weights=False, const_init=True, normalize=False, use_static_covariates=True)
modelnl2.fit(train2, verbose = True)
prednl2 = modelnl2.predict(n = len(val2), series = train2)

modelnl3 = NLinearModel(input_chunk_length=24 , output_chunk_length=12, shared_weights=False, const_init=True, normalize=False, use_static_covariates=True)
modelnl3.fit(train3, verbose = True)
prednl3 = modelnl3.predict(n = len(val3), series = train3)
modelnl4 = NLinearModel(input_chunk_length=24 , output_chunk_length=12, shared_weights=False, const_init=True, normalize=False, use_static_covariates=True)
modelnl4.fit(train4, verbose = True)
prednl4 = modelnl4.predict(n = len(val4), series = train4)

modelnl7 = NLinearModel(input_chunk_length=24 , output_chunk_length=12, shared_weights=False, const_init=True, normalize=False, use_static_covariates=True)
modelnl7.fit(train7, verbose = True)
prednl7 = modelnl7.predict(n = len(val7), series = train7)

modelnl8 = NLinearModel(input_chunk_length=24 , output_chunk_length=12, shared_weights=False, const_init=True, normalize=False, use_static_covariates=True)
modelnl8.fit(train8, verbose = True)
prednl8 = modelnl8.predict(n = len(val8), series = train8)
```

```python
import pandas as pd
from darts.metrics import rmse, mae, mape

# Create a DataFrame to store the evaluation results
evaluation_results = pd.DataFrame(columns=['Target', 'RMSE', 'MAE', 'MAPE'])

# Calculate RMSE and MAE for target1
rmse1 = rmse(val1, prednl1)
mae1 = mae(val1, prednl1)
mape1 = mape(val1, prednl1)
evaluation_results = evaluation_results.append({'Target': 'target1', 'RMSE': rmse1, 'MAE': mae1,'MAPE': mape1}, ignore_index=True)

# Calculate RMSE and MAE for target2
rmse2 = rmse(val2, prednl2)
mae2 = mae(val2, prednl2)
mape2 = mape(val2, prednl2)
evaluation_results = evaluation_results.append({'Target': 'target2', 'RMSE': rmse2, 'MAE': mae2,'MAPE': mape2}, ignore_index=True)


# Calculate RMSE and MAE for target3
rmse3 = rmse(val3, prednl3)
mae3 = mae(val3, prednl3)
mape3 = mape(val3, prednl3)
evaluation_results = evaluation_results.append({'Target': 'target3', 'RMSE': rmse3, 'MAE': mae3,'MAPE': mape3}, ignore_index=True)


# Calculate RMSE and MAE for target4
rmse4 = rmse(val4, prednl4)
mae4 = mae(val4, prednl4)
mape4 = mape(val4, prednl4)
evaluation_results = evaluation_results.append({'Target': 'target4', 'RMSE': rmse4, 'MAE': mae4,'MAPE': mape4}, ignore_index=True)


# Calculate RMSE and MAE for target7
rmse7 = rmse(val7, prednl7)
mae7 = mae(val7, prednl7)
mape7 = mape(val7, prednl7)
evaluation_results = evaluation_results.append({'Target': 'target7', 'RMSE': rmse7, 'MAE': mae7,'MAPE': mape7}, ignore_index=True)


# Calculate RMSE and MAE for target8
rmse8 = rmse(val8, prednl8)
mae8 = mae(val8, prednl8)
mape8 = mape(val8, prednl8)
evaluation_results = evaluation_results.append({'Target': 'target8', 'RMSE': rmse8, 'MAE': mae8,'MAPE': mape8}, ignore_index=True)


# Save the evaluation results to a CSV file
evaluation_results.to_csv('Nlinear_2.csv', index=False)
```
